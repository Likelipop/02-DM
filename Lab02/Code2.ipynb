# PROBLEM 1

```python
# Take the values
array = df.values
array

# Initialize the Similarity Matrix
s1_matrix = np.zeros((50,50))
s2_matrix = np.zeros((50, 50))
sinf_matrix = np.zeros((50, 50))

#Metric function
def metric(x,y,p=2):
    norm = 0
    if p == 1:
        norm = np.sum(np.abs(x-y))
    elif p == 2:
        norm = np.sqrt(np.sum( np.square(x-y)))
    elif p == "inf":
        norm =np.max(np.abs(x,y))
    
    return norm
```

<h4> P = 1 </h4>

```python
for i,x in enumerate(array[:50]):
    for j,y in enumerate(array[:50]):
        dist_xy = metric(x,y,1)
        s1_matrix[i,j]  = dist_xy 

s1_df = pd.DataFrame(s1_matrix)
s1_df.head()
```

<h4> P = 2 </h4>

```python
for i,x in enumerate(array[:50]):
    for j,y in enumerate(array[:50]):
        dist_xy = metric(x,y,2)
        s2_matrix[i,j]  = dist_xy 

s2_df = pd.DataFrame(s2_matrix)
s2_df.head()
```

<h4> P = inf </h4>

```python
for i,x in enumerate(array[:50]):
    for j,y in enumerate(array[:50]):
        dist_xy = metric(x,y,"inf")
        sinf_matrix[i,j]  = dist_xy 

sinf_df = pd.DataFrame(sinf_matrix)
sinf_df.head()
```

<h3>4. create 2 distance measure function </h3>

```python
from collections import defaultdict

def overlap_similarity(x,y) -> float:
    d = x.shape[0]
    return ((x == y).sum())*float(1/d)
    
def invertf_similarity(x,y) -> float:
    d = x.shape[0]
    f_x = defaultdict(int)
    f_y = defaultdict(int)
    for item in x:
        f_x[item] +=1
    for item in y:
        f_y[item] +=1

    value = 0
    for x_item,y_item in zip(x,y):
        if x_item == y_item:
            value+= 1
            next
        value+= float(1/(1+ np.log(f_x[x_item] + f_y[y_item])))

    return value/float(d)
```

```python
overlapf_matrix = np.zeros((100,100))
invertf_matrix = np.zeros((100,100))
```

```python
for i,x in enumerate(categorical_data.values[:100]):
    for j,y in enumerate(categorical_data.values[:100]):
        dist_xy = overlap_similarity(x,y)
        overlapf_matrix[i,j]  = dist_xy

overlapf_df = pd.DataFrame(overlapf_matrix)
overlapf_df.head()
```

```python
for i,x in enumerate(categorical_data.values[:100]):
    for j,y in enumerate(categorical_data.values[:100]):
        dist_xy = invertf_similarity(x,y)
        overlapf_matrix[i,j]  = dist_xy

overlapf_df = pd.DataFrame(overlapf_matrix)
overlapf_df.head()
```
